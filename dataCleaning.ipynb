```python
# import dependencies
import pandas as pd
```

```python
# import raw data for june 2019
dfJun19 = pd.read_csv('Resources/2019/201906-citibike-tripdata.csv')
dfJun19.head()
```

```python
# import raw data for june 2020
dfJun20 = pd.read_csv('Resources/2020/202006-citibike-tripdata.csv')
dfJun20.head()
```

```python
dfJun18 = pd.read_csv('Resources/June/201806-citibike-tripdata.csv')
dfJun18.head()
```

```python
dfJun17 = pd.read_csv('Resources/June/201706-citibike-tripdata.csv')
dfJun17.head()
```

```python
# print size of raw dataframes
print(len(dfJun17.index))
print(len(dfJun18.index))
print(len(dfJun19.index))
print(len(dfJun20.index))
```

```python
# drop na rows from raw data
dfJun17 = dfJun17.dropna()
dfJun18 = dfJun18.dropna()
dfJun19 = dfJun19.dropna()
dfJun20 = dfJun20.dropna()
```

```python
# generate lists of station info column names
startList = ["start station id", "start station name", "start station latitude", "start station longitude"]
endList = ["end station id", "end station name", "end station latitude", "end station longitude"]
```

```python
# pull station info into separate dataframes, drop duplicates
startStations17 = dfJun17[startList].drop_duplicates(subset="start station id")
startStations18 = dfJun18[startList].drop_duplicates(subset="start station id")
startStations19 = dfJun19[startList].drop_duplicates(subset="start station id")
startStations20 = dfJun20[startList].drop_duplicates(subset="start station id")

endStations17 = dfJun17[endList].drop_duplicates(subset="end station id")
endStations18 = dfJun18[endList].drop_duplicates(subset="end station id")
endStations19 = dfJun19[endList].drop_duplicates(subset="end station id")
endStations20 = dfJun20[endList].drop_duplicates(subset="end station id")
```

```python
# print length of each dataframe
print(f"unique start stations 2017: {len(startStations17.index)}")
print(f"unique end stations 2017: {len(endStations17.index)}")
print(f"unique start stations 2018: {len(startStations18.index)}")
print(f"unique end stations 2018: {len(endStations18.index)}")
print(f"unique start stations 2019: {len(startStations19.index)}")
print(f"unique end stations 2019: {len(endStations19.index)}")
print(f"unique start stations 2020: {len(startStations20.index)}")
print(f"unique end stations 2020: {len(endStations20.index)}")
```

```python
# generate list of column names for combined station table
stationCols = ["Station ID", "Station Name", "Station Latitude", "Station Longitude"]
```

```python
# set column names for all 4 dataframes to be the same
startStations17.columns = stationCols
startStations18.columns = stationCols
startStations19.columns = stationCols
startStations20.columns = stationCols

endStations17.columns = stationCols
endStations18.columns = stationCols
endStations19.columns = stationCols
endStations20.columns = stationCols
```

```python
# concatenate all 4 station dfs
frames = [startStations17, startStations18, startStations19, startStations20, 
          endStations17, endStations18, endStations19, endStations20]

concatStationsDf = pd.concat(frames)

concatStationsDf
```

```python
# drop duplicate rows by stations id
station_df = concatStationsDf.drop_duplicates(subset="Station ID")
station_df
```

```python
# add second column of station ids to match with other tables better
station_df["end station id"] = station_df["Station ID"]
station_df
```

```python
# redo column names to better match raw data column names
redoCols = ["start station id", "Station Name", "Station Latitude", "Station Longitude", "end station id"]
station_df.columns = redoCols
```
